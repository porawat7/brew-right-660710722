## แสดง ปี ประเภท ชนิด เกษตรกร พื้นที่

```python
query = """
  SELECT f.ปี, f.ประเภท, f.ชนิด,เกษตรกร, พื้นที่
  FROM 'Farmer.csv' f RIGHT JOIN 'BigField.csv' b
  ON f.ปี = b.ปี  AND f.ประเภท = b.ประเภท
"""
```

```python
Q3 = con.execute(query).fetchdf()
```

```python
Q3
```

## แสดง ปี ประเภท ชนิด ขนาดของพื้นที่(โดยประมาณ)

```python
query = """
  SELECT ปี, ประเภท, ชนิด,
  CASE
  WHEN TRY_CAST(พื้นที่ as float) > 1000 then 'ใหญ่มาก'
  WHEN TRY_CAST(พื้นที่ as float) >= 500 AND TRY_CAST(พื้นที่ as float) < 1000  then 'ใหญ่'
  ELSE 'ค่อนข้างใหญ่'
  END AS ขนาด
  FROM 'BigField.csv'
"""
```

```python
Q4 = con.execute(query).fetchdf()
```

```python
Q4
```

## แสดง  ปี ประเภท ชนิด เกษตรกร   จำนวนเกษตรกรที่ปลูกพืชแต่ละประเภทมากที่สุดในปี_2559

```python
query = """
select
 ปี,ประเภท,ชนิด,เกษตรกร as จำนวนเกษตรกรที่ปลูกพืชแต่ละประเภทมากที่สุดในปี_2559
 from 'Farmer.csv'
 WHERE ปี = 2559
 QUALIFY ROW_NUMBER() OVER (PARTITION BY ประเภท ORDER BY เกษตรกร DESC) = 1
"""
```

```python
Q5 = con.execute(query).fetchdf()
```

```python
Q5
```

## แสดง  ปี ประเภท ชนิด เกษตรกร  จำนวนเกษตรกรที่ปลูกพืชแต่ละประเภทน้อยที่สุดในปี 2561

```python
query = """
select
 ปี,ประเภท,ชนิด,เกษตรกร as จำนวนเกษตรกรที่ปลูกพืชแต่ละประเภทน้อยที่สุด
 from 'Farmer.csv'
 WHERE ปี = 2561
 QUALIFY ROW_NUMBER() OVER (PARTITION BY ประเภท ORDER BY เกษตรกร ASC) = 1
"""
```

```python
Q6 = con.execute(query).fetchdf()
```

```python
Q6
```

## แสดง ปี ประเภท จำนวนเกษตรกรและ จำนวนประเภทของพืชที่ปลูกแต่ละปี

```python
query = """
select
 ANY_VALUE(ปี) as ปี,ประเภท,SUM(เกษตรกร) as จำนวนเกษตรกร,COUNT(ประเภท) as จำนวนประเภทของพืชที่ปลูกแต่ละปี
 from 'Farmer.csv'
 group by ประเภท
 order by ANY_VALUE(ปี)
"""
```

```python
Q7 = con.execute(query).fetchdf()
```

```python
Q7
```

## แสดง ปี ประเภท จำนวนชนิดของพืชที่ปลูกแต่ละปี

```python
query = """
select
 ANY_VALUE(ปี) as ปี,ประเภท,COUNT(ชนิด) as จำนวนชนิดของพืชที่ปลูกแต่ละปี
 from 'Farmer.csv'
 group by ประเภท
 order by ANY_VALUE(ปี), จำนวนชนิดของพืชที่ปลูกแต่ละปี DESC
"""
```

```python
Q8 = con.execute(query).fetchdf()
```

```python
Q8
```

## แสดง ปี ประเภท จำนวนชนิด จำนวนเกษตรกรต่อปี

```python
query = """
select
 ANY_VALUE(ปี) as ปี,ประเภท,COUNT(ชนิด) as จำนวนชนิด,sum(เกษตรกร) as จำนวนเกษตรกรต่อปี
 from 'Farmer.csv'
 group by ประเภท
 order by ANY_VALUE(ปี), จำนวนเกษตรกรต่อปี
"""
```

```python
Q9 = con.execute(query).fetchdf()
```

```python
Q9
```

## แสดง ชนิด ประเภท และ ขนาดพื้นที่การเกษตรแต่ละชนิด

```python
query = """
select
 ชนิด,ANY_VALUE(ประเภท) as ประเภท,sum(TRY_CAST(พื้นที่ as double)) as พื้นที่การเกษตรแต่ละชนิด
 FROM 'BigField.csv'
 group by ชนิด
 order by พื้นที่การเกษตรแต่ละชนิด
"""
```

```python
Q10 = con.execute(query).fetchdf()
```
